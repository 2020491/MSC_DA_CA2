# Initial Cleaning


## Remove Irrelevant Data
To reduce the size of the data, I will remove any irrelevant data (for the analysis) from the datasets.

### `Flag`

```python
# -- TRADE
#df_trade.Flag
#df.drop(df.filter(like='Code').columns, 1)
df_trade = df_trade[df_trade.columns[~df_trade.columns.str.contains('Flag')]]

# -- PRICE
df_price = df_price[df_price.columns[~df_price.columns.str.contains('Flag')]]
```

### `Year Code`
The `Year Code` column is identical with `Year` column and so can be removed (with its duplicate data).

```python
# -- TRADE
print("-- TRADE DATA  ------------------------------")
print("Columns Same? "+str(df_trade['Year'].equals(df_trade['Year Code'])))

# -- PRICE
print("\n-- PRICE DATA ------------------------------")
print('Columns Same? ' + str(df_price['Year'].equals(df_price['Year Code'])))
```

```python
# -- TRADE
#df_trade.drop(df_trade.filter(like='Year Code').columns, 1)
df_trade.drop('Year Code', inplace=True, axis=1)

# -- PRICE
df_price.drop('Year Code', inplace=True, axis=1)
```

### `Element Code`
As seen below `Element Code` is unusable with multiple values for the same `Element`, for instance (in the Trade Data) `Export Quantity` is represented by *5907*, *5908*, *5909*, *5910*. To filter the type of Trade for different situations it is enough to use `Element` and drop the inconsistent `Element Code`.    


```python
# -- TRADE
print("-- TRADE DATA  ------------------------------")
for value in df_trade.sort_values("Element Code")['Element Code'].unique():
    print(str(value)+":"+str(list(df_trade.loc[df_trade['Element Code'].isin([value])].Element.unique())))
    
# -- PRICE
print("\n-- PRICE DATA ------------------------------")
for value in df_price.sort_values("Element Code")['Element Code'].unique():
    print(str(value)+":"+str(list(df_price.loc[df_price['Element Code'].isin([value])].Element.unique())))      
```

```python
# -- TRADE
df_trade.drop('Element Code', inplace=True, axis=1)

# -- PRICE
df_price.drop('Element Code', inplace=True, axis=1)
```

### `Partner`, `Reporter` & `Area` `Country` `Code (M49)` & `Item Code (CPC)`
In the Trade and Price Data, there are specially encryped codes which are unnecessary, due to having another easily useable representative for them, for instance (in Price Data), `Area Code` and `Area Code (M49)` depict the same data in different format. Therefore, they are removed from the datasets.

```python
# -- TRADE
df_trade.drop(['Reporter Country Code (M49)','Partner Country Code (M49)','Item Code (CPC)'], inplace=True, axis=1)

# -- PRICE
df_price.drop(['Area Code (M49)','Item Code (CPC)'], inplace=True, axis=1)
```

### `Months Code`
As seen below, `Months Code` sorts the data based on what month it is. So for the initial cleaning I will leave it in. May remove later...

```python
 # -- PRICE
print("\n-- PRICE DATA ------------------------------")
for value in df_price.sort_values("Months Code")['Months Code'].unique():
    print(str(value)+":"+str(list(df_price.loc[df_price['Months Code'].isin([value])].Months.unique())))      
```

### `Unit`
Still could be useful for later analysis, to clarify the specific units represented in the trading and price data. Although trends can still be explored without this data.

```python
# -- TRADE
print("-- TRADE DATA  ------------------------------")
#display(list(df_trade.Unit.unique()))
for value in df_trade.sort_values("Unit")['Unit'].unique():
    print(str(value)+":"+str(list(df_trade.loc[df_trade['Unit'].isin([value])].Element.unique())))

# -- PRICE
print("\n-- PRICE DATA ------------------------------")
#display(list(df_price.Unit.unique()))
for value in df_price.sort_values("Unit")['Unit'].unique():
    print(str(value)+":"+str(list(df_price.loc[df_price['Unit'].isin([value])].Element.unique())))

#print("Columns Same? "+str(df_trade['Year'].equals(df_trade['Year Code'])))
#print('Columns Same? ' + str(df_price['Year'].equals(df_price['Year Code'])))
```

## Check for Missing Value
Trading Data that has missing `Value` are unusable and as such are dropped from the anaysis.<br>
While Price Data having `Unit` data missing is irrelevant especially since this missing data appear when the `Element` column is **Producer Price Index (2014-2016 = 100)** as can be seen above.

```python
# ANY MISSING VALUES?

# -- TRADE
print("-- TRADE DATA  ------------------------------")
print("Missing Value? "+str(df_trade.isnull().values.any()))

# -- PRICE
print("\n-- PRICE DATA ------------------------------")
print('Missing Value? ' + str(df_price.isnull().values.any()))      
```

```python
# COUNT NUMBER OF NULL/NaN VALUES IN EACH COLUMN

# -- TRADE
print("-- TRADE DATA  ------------------------------")
display(df_trade.isnull().sum())

# -- PRICE
print("\n-- PRICE DATA ------------------------------")
display(df_price.isnull().sum())
```

```python
# Data Filtering: displaying data only with Variable = NaN
# -- TRADE
df_trade[df_trade.Value.isnull()]
```

```python
# DROP ROWS WITH MISSING VALUES (CAREFUL WITH MEMORY USAGE --> MemoryError: Unable to allocate X MiB for an array with shape (X,Y) and data type X)

# -- TRADE
df_trade.dropna(inplace=True)
df_trade.reset_index(drop=True, inplace=True)
```

```python
# verification
df_trade[df_trade.Value.isnull()]
```

## Check for Duplicate Data
There are no duplicate rows of Data (i.e. records) in either Trade or Price.

```python
# -- TRADE
print("-- TRADE DATA  ------------------------------")
print("Duplicate Value? "+str('Yes' if len(df_trade[df_trade.duplicated()])>0 else "No"))

# -- PRICE
print("\n-- PRICE DATA ------------------------------")
print("Duplicate Value? "+str('Yes' if len(df_price[df_price.duplicated()])>0 else "No"))
```

## Remove Irrelevant Records
### `Reporter Countries` & `Area`
The analysis is conducted on specific countries in mind. So I will filter for these specific countries as `Reporter Countries` (in Trade) and `Area` (in Price).

```python
# Trade Analysis: Filter for Specific counties
df_trade = df_trade[df_trade["Reporter Countries"].isin(["Ireland", "Belgium","France","Netherlands", "Germany"])].reset_index(drop=True)
#df_trade.head()

# Price Analysis: Filter for Specific Countries
df_price = df_price[df_price["Area"].isin(["Ireland", "Belgium","France","Netherlands", "Germany"])].reset_index(drop=True)
#df_price.head()
```
